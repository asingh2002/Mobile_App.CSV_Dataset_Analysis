Convert the values of Android_Ver as

*   '4.0.3' -> 4.0

    '4.4' -> 4.4



```python
re.sub(' .*', '', "4.0.3 and up")
```

```python
l=[]
for i in new:
  if i != "Varies with device":
    a=re.sub(' .*', '', str(i))
    l.append(a)
  elif i == "Varies with device":
    l.append("Varies with device")
```

```python
test = pd.DataFrame(l, columns = ["Android_Ver"])
test.value_counts()
```

```python
# Based on the above hypothesis we can replace the value "Varies with device" to "4.1"
test = test.replace("Varies with device", "4.1")
```
